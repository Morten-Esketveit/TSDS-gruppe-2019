```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
```

#### Producing maps using spatial information

Load data. Note, that TSDS_DataProcessing must be run to create .pkl files.

```python
airports = pd.read_pickle("airports.pkl")
connections = pd.read_pickle("connections.pkl")
```

#### Load world map

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world = world[(world.pop_est>0) & (world.name!="Antarctica")]
```

#### Plot maps 

Note, that the 'Connected' indicator shows whether the airport is found in the connections dataframe - quite a few airports are not, and I suspect they are of lesser importance. 

#### Plot map with all airports (connected and not)

```python
fig, ax = plt.subplots(figsize = (24,24))
world.plot(ax = ax, color = "green", edgecolor = "black")
airports.loc[airports.Connected == 1].plot(ax = ax, color = "cyan", markersize = 2, marker = "o")
airports.loc[airports.Connected == 0].plot(ax = ax, color = "darkblue", markersize = 2, marker = "x")
ax.set_facecolor("lightblue")
plt.xticks([])
plt.yticks([])
plt.savefig(fname = "Airports_WorldMap.pdf", format = "pdf")
plt.show()
```
